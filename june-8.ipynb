**Q11)**

**1. User have to create a new password in starting time (user input).**

**2. User again give the password for checking.**

**3. If it will give 3 times for re-entering the password.**

**4. After that it will display "Exceeded maximum limit Access denied."**

```python
pw = input("Create your password:")
for i in range(3,0,-1):
    if pw == input("\nRe-enter your password:"):
        print("Login successful")
        break
    else: print(i-1, "chances left")
else: print("\nExceeded maximum limit Access denied.")
```

```python
#WAP to create a function called calculator for the calculation of addition, subraction, multiplication and division of two numbers where user give two
#number with the operatot
def calculator(n1, n2, op):
    if op == "+": print("Result:",n1+n2)
    elif op == "-": print("Result:",n1-n2)
    elif op == "*": print("Result:",n1*n2)
    elif op == "/": print("Result:",n1/n2)
    else: return "Invalid Operator"
calculator(int(input("Number 1:")),int(input("Number 2:")),input("Operator:"))
```

```python
def calculator(n1, n2, op):
    try:
        if op == "+": print("Result:",n1+n2)
        elif op == "-": print("Result:",n1-n2)
        elif op == "*": print("Result:",n1*n2)
        elif op == "/": print("Result:",n1/n2)
        else: return "Invalid Operator"
    except ZeroDivisionError: print("Error!!! Division by zero!")
calculator(int(input("Number 1:")),int(input("Number 2:")),input("Operator:"))
```

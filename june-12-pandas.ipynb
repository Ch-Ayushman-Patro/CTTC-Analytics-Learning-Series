# Auto MPG Data Analysis

```python
adf = pd.read_csv(r".\Datasets\Auto-MPG-Dataset\auto-mpg.data", delim_whitespace = True, header = None)
adf
```

```python
adf.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model_year', 'origin', 'car_name']
adf
```

```python
adf.isnull()
```

```python
adf.isnull().sum()
```

```python
adf.dtypes
```

```python
for i in adf.columns: print(i,":\n",adf[i].unique())
```

```python
adf['horsepower'].unique()
```

```python
for i in ['horsepower', 'car_name']: print(i, ':"', (adf[i] == '?').sum())
```

```python
for i in ['horsepower', 'car_name']: print(i, ':"', sum(adf[i] == '?'))
```

```python
adf['horsepower'].to_string().count('?')
```

```python
adf.horsepower.mode()
```

```python
for i in ['horsepower']: adf.replace(['?'], adf.horsepower.mode(), inplace = True)
```

```python
adf.horsepower.to_string().split("\n")
```

```python
adf['horsepower'].to_string().count('?')
```

```python
adf.horsepower = adf.horsepower.astype(float)
adf.dtypes
```

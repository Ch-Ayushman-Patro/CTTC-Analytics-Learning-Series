**Create a Bird class, where bird name, age, color and weight are class attributes and it have a eat method which will contain a parameter amount and will return the total weightof the bird by adding the amount of food it has eaten and its weight**

```python
class Bird:
    name = 'Ashok'
    age = 20
    color = 'Red'
    weight = 3
    def eat(self, amount):
        return self.weight + amount
b = Bird()
b.eat(0.8)
```

# Lambda Function

```python
def isEven(x):
    if x % 2 == 0:
        print('Even')
    else:
        print('Odd')
isEven(20)
```

```python
isEven = lambda x : print('Even') if x % 2 == 0 else print('Odd')
isEven(20)
isEven(15)
```

```python
c = input("Enter anything: ")
if c == c[::-1]:
    print("Palindrome")
else:
    print("Not Palindrome")
```

```python
pd = lambda x : print('Palindrome') if x == x[::-1] else print('Not Palindrome')
pd(input("Enter anything: "))
```

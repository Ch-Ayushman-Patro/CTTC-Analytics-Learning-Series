```python
ip = df1.drop(['Anomaly Scores'], axis = 1)
ip
```

```python
op = df1['Anomaly Scores']
op
```

```python
x_train, x_test, y_train, y_test = train_test_split(ip, op, train_size = 0.8)
df.shape, x_train.shape, x_test.shape
```

```python
sc = StandardScaler()
sc
```

```python
x_train = sc.fit_transform(x_train)
x_test = sc.fit_transform(x_test)
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
ycap = lr.predict(x_test)
ycap
```

```python
pdf = pd.DataFrame({'y':list(y_test),
                    'prediction':ycap})
pdf
```

```python
mse = mean_squared_error(y_test, ycap)
print("Mean Squared Error: ", mse)
```

```python
r2 = r2_score(y_test, ycap)
print("R2 Score:", r2)
print("Accuracy:", r2*100)
```

```python
sns.kdeplot(df1['Anomaly Scores'], label = "Actual Value")
sns.kdeplot(y_test, label = "Testing Values")
sns.kdeplot(ycap, label = "Predicted Values")
plt.legend()
plt.show()
```

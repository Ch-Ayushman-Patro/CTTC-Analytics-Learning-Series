# Distribution Plot

*sns.distplot( < data >[, color = < color_code >, label = < label_name >])*

```python
sns.distplot(idf.SL)
plt.title('Distribution of SL (Sepal Length)')
plt.show()
```

```python
sns.distplot(idf.SL[idf.CLASS == 'Iris-setosa'], label = 'Iris-setosa')
sns.distplot(idf.SL[idf.CLASS == 'Iris-virginica'], label = 'Iris-virginica')
sns.distplot(idf.SL[idf.CLASS == 'Iris-versicolor'], label = 'Iris-versicolor')
plt.title('Distribution of SL (Sepal Length) w.r.t CLASS')
plt.legend()
plt.show()
```

# Dis Plot

```python
sns.displot(idf.SL)
plt.show()
```

```python
sns.displot(idf.SL, fill = False, hatch = '*', color = 'orange')
plt.show()
```

```python
sns.displot(idf.SL, fill = False, hatch = '/', color = 'red')
plt.show()
```

```python
sns.displot(idf.SL, hatch = '/', color = 'red')
plt.show()
```

# KDE Plot -> distplot without histogram

*sns.kdeplot( < data >[, color = < color_code >, label = < label_name >, fill = < True / False >, hatch = < hatch_type >])*

```python
sns.kdeplot(idf.SL, color = 'g')
plt.title('Distribution curve of SL (Sepal Length)')
plt.show()
```

```python
sns.kdeplot(idf.SL[idf.CLASS == 'Iris-setosa'], label = 'Iris-setosa')
sns.kdeplot(idf.SL[idf.CLASS == 'Iris-virginica'], label = 'Iris-virginica')
sns.kdeplot(idf.SL[idf.CLASS == 'Iris-versicolor'], label = 'Iris-versicolor')
plt.title('Distribution curve of SL (Sepal Length) w.r.t CLASS')
plt.legend()
plt.show()
```

```python
sns.kdeplot(idf.SL[idf.CLASS == 'Iris-setosa'], fill = True, hatch = '/', label = 'Iris-setosa')
sns.kdeplot(idf.SL[idf.CLASS == 'Iris-virginica'], fill = True, hatch = '/', label = 'Iris-virginica')
sns.kdeplot(idf.SL[idf.CLASS == 'Iris-versicolor'], fill = True, hatch = '/', label = 'Iris-versicolor')
plt.title('Distribution curve of SL (Sepal Length) w.r.t CLASS')
plt.legend()
plt.show()
```

# Correlation Plot using Heatmap

*sns.heatmap( < data >[, annot = True, cmap = < color_map_code >] )*

where,

- data must be numerical or integer data
- annot refers to the inclusion of numericl value
- cmap refer to color coding of lowest and highest values accordingly

```python
idf1 = idf.drop('CLASS', axis = 1)
idf1
```

```python
corr_idf = idf1.corr()
corr_idf
```

```python
sns.heatmap(corr_idf)
plt.show()
```

```python
sns.heatmap(corr_idf, annot = True)
plt.show()
```

```python
sns.heatmap(corr_idf, annot = True, cmap = 'coolwarm')
plt.show()
```

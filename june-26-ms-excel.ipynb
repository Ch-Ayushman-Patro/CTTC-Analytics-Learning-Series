# Excel Commands

+ **Shift + Space** -> Select Row
+ **Ctrl + Space** -> Select Column
+ **Ctrl + Shift + Space** -> New Row
+ **Ctrl + -** -> Delete
+ **Ctrl + A** -> Selects the whole worksheet
+ **Ctrl + Enter** -> Applies the operation of a single cell on the whole selection

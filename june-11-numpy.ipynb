# Array Indexing and Slicing

```python
a3 = np.array([1, 2, 3, 4])
a3
```

```python
a3[0]
```

```python
a3[1:3]
```

```python
m2 = np.array([[2, 3, 4], [8, 4, 5], [6, 2, 4]])
m2
```

In 2-dimensional array, the **indexing** goes by first **ROW** and then **COLUMN** => [< row >, < column >]

```python
m2[0,0]
```

```python
m2[2,2]
```

```python
m2[2,0]
```

```python
np.where(m2 == 8)
```

```python
m3 = np.array([[2, 5, 6, 8], [1, 2, 3, 9], [7, 8, 4, 5], [6, 6, 2, 3]])
m3
```

In 2-dimensional array, the **sliciing** goes by first **ROW SLICING** and then **COLUMN SLICING** => [< row_slice >, < column_slice >]

```python
m3[1:3,1:3]
```

```python
m3[0:2,1:3]
```

```python
m3[:2,1:3]
```

```python
m3[1:3,1:5]
```

```python
m3[1:3,1:]
```

```python
n = int(input("Enter a number:"))
if n>0:
    print("Positive")
elif n==0:
    print("Zero(0)")
else:
    print("Negative")
```

```python
n = int(input("Enter a number:"))
if n%2==0 and n>0: print("Even and Positive")
elif n%2==0 and n<0: print("Even and Negative")
elif n%2!=0 and n>0: print("Odd and Positive")
else: print("Odd and Negative")
```

```python
# WAP to check the grades obtained if user input percentage of marks
# > 90 - A
# < 90 and > 80 - B
# < 80 and > 60 - C
# Below 60 - D
m = int(input())
if m>90: print("A")
elif m>80 and m<90: print("B")
elif m>60 and m<80: print("C")
else: print("D")
```

```python
import math as m
n, c = int(input()), 0
for i in range(2,int(m.sqrt(n))):
    if n%i == 0: c+=1
if n == 1: print("Neither a Composite nor a Prime")
elif c == 0: print("Prime")
else: print("Composite")
```

```python
n = int(input())
c = 0
for i in range(2,n):
    if n%i == 0:
        c+=1
if n == 1:
    print("Neither a Composite nor a Prime")
elif c == 0:
    print("Prime")
else:
    print("Composite")
```

```python
n = int(input())
if n == int(str(n)[::-1]): print("Palindrome")
else: print("Not Palindrome")
```

```python
n1 = n2 = int(input())
s = 0
while n1>0:
    i = n1%10
    s = s*10 + i
    n1//=10
if n2 == s: print("Palindrome")
else: print("Not Palindrome")
```

```python
#WAP to check whether a number is Adom number or not
import math as m
x = int(input())
y = int(str(int(m.sqrt(int(str(x**2)[::-1]))))[::-1])
if x == y: print("Adom Number")
else: print("Not Adom Number")
```

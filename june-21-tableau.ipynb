# Tableau

**5 Products of Tableau :**
1. Tableau Desktop
2. Tableau Online
3. Tableau Server
4. Tableau Public
5. Tableau Reader

# Sort Fields

1. **A to Z ascending :** Sorts the 'Columns' in Ascending order
2. **Z to A descending :** Sorts the 'Columns' in Descending order
3. **A to Z ascending per table :** Sorts the 'Columns' in Ascending order in consideration of multiple tables
4. **Z to A descending per table :** Sorts the 'Columns' in Descending order in consideration of multiple tables

# Datatypes

1. Number ( decimal )
2. Number ( whole )
3. Date & Time
4. Date
5. String
6. Boolean
7. Geographical Role

# Connections

+ **Live -** Changes on timely made on dataset will reflect on Tableau
+ **Extract -** Changes on timely made on dataset will not reflect on Tableau
